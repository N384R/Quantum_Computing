1. HEA depth 4

```python
from pyscf import gto
from qc_practice import VQE, SSVQE
from qc_practice.ansatz import UCCSD, SpinFlipUCCSD, HEA, UCCGSD, SP

mol = gto.M(atom = 'H 0 0 0; H 0 0 2.30', basis = 'sto-3g')
ssvqe = SSVQE(mol)
ssvqe.ansatz = HEA(depth=4)
ssvqe.parallel = True
ssvqe.run()
```
